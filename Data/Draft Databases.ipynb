```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/2010_NFL_Draft').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find("table",{'class':'wikitable sortable'})
My_table
```

```python
links = My_table.findAll('a')
links
```

```python
Players = []
for link in links:
    Players.append(link.get('title'))
    
print((Players))
```

```python
import pandas as pd
df = pd.DataFrame(columns = ['Team', 'Player', 'Position'])
```

```python
for element in Players:
    if (element == None):
        Players.remove(element)
del Players[336:338]
Players
```

```python

for ind, obj in enumerate(Players):
    if (ind % 5 == 3):
        to_append = [Players[ind-2], Players[ind-1], Players[ind]]
        df_length = len(df)
        df.loc[df_length] = to_append
    
```

```python
df
df.drop([509], inplace = True)
```
